# Fix daily attributes

```python
from day_info import get_daily_info_df_for_dates_and_city
```

```python
start = pd.Timestamp('2022-01-01')
end = pd.Timestamp('2022-12-31')
dates_2022 = pd.date_range(start, end)
```

```python
daily = get_daily_info_df_for_dates_and_city(dates_2022,"Brussels", "Belgium")
```

```python
daily.drop(columns=['maxTempC'], inplace=True)
daily.drop(columns=['minTempC'], inplace=True)
daily.drop(columns=['avgTempC'], inplace=True)
daily.drop(columns=['sunHour'], inplace=True)
daily.drop(columns=['uvIndex'], inplace=True)
daily.drop(columns=['feelsLikeC'], inplace=True)
daily.drop(columns=['windspeedKmph'], inplace=True)
daily.drop(columns=['humidity'], inplace=True)

```

```python
from datetime import datetime 
from meteostat import Point, Daily, units  
# Set time period   
# Create Point for DC 
location = Point(50.8465573, 4.351697)  
# Get daily data 
weather_data = Daily(location, start, end) 
weather_data = weather_data.fetch()
weather_data.drop(columns=['snow'], inplace=True)
weather_data.drop(columns=['tsun'], inplace=True)
weather_data.head()
```

```python
daily_merged = daily.merge(weather_data, left_index=True, right_index=True, how="left")
```

```python
import pandas as pd

# Assuming your existing DataFrame is named 'existing_df'
# And you want to repeat it 100 times

# Repeat the existing DataFrame 100 times
repeated_df = pd.concat([daily_merged] * 1300, ignore_index=True)

# Create meterid index repeating from 1 to 100
meterid = sum([[i]*len(daily) for i in range(1, 1301)], [])

# Set meterid as the index of the repeated DataFrame
repeated_df.index = pd.Index(meterid, name='meterID')

repeated_df.head()
```

```python
userattributes = data.groupby('EAN_ID').max()[['Warmtepomp_Indicator', 'Elektrisch_Voertuig_Indicator', 'PV-Installatie_Indicator']]

# Reset index to make EAN_ID a column again
userattributes.reset_index(inplace=True)
userattributes.head()
```

```python
warmtepomp_values = []

# Loop over each value in the 'Warmtepomp_Indicator' column
for value in userattributes['Warmtepomp_Indicator']:
    # Append the value to the list 365 times
    warmtepomp_values.extend([value] * 365)
    
Elektrisch_Voertuig_Indicator_values = []

# Loop over each value in the 'Warmtepomp_Indicator' column
for value in userattributes['Elektrisch_Voertuig_Indicator']:
    # Append the value to the list 365 times
    Elektrisch_Voertuig_Indicator_values.extend([value] * 365)
    
PV_Installatie_Indicator_values = []

# Loop over each value in the 'Warmtepomp_Indicator' column
for value in userattributes['PV-Installatie_Indicator']:
    # Append the value to the list 365 times
    PV_Installatie_Indicator_values.extend([value] * 365)
```

```python
repeated_df['Warmtepomp_Indicator'] = warmtepomp_values
repeated_df['Elektrisch_Voertuig_Indicator'] = Elektrisch_Voertuig_Indicator_values
repeated_df['PV-Installatie_Indicator'] = PV_Installatie_Indicator_values
```

```python

# Group by 'EAN_ID' and sum the 'VolumeAfname_TH_kWh' and 'VolumeInjectie_TH_kWh' columns
	
grouped_df = data.groupby('EAN_ID').agg({'Volume_Afname_kWh': 'sum', 'Volume_Injectie_kWh': 'sum'})

# Calculate the yearly consumption
grouped_df['yearly_consumption'] = grouped_df['Volume_Afname_kWh'] - grouped_df['Volume_Injectie_kWh']
grouped_df['injected'] = grouped_df['Volume_Injectie_kWh']
grouped_df['consumed'] = grouped_df['Volume_Afname_kWh']

# Drop unnecessary columns
grouped_df.drop(['Volume_Afname_kWh', 'Volume_Injectie_kWh'], axis=1, inplace=True)
grouped_df = grouped_df.rename_axis('meterID')

grouped_df.head()
```
