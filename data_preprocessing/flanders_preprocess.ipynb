```python
import pandas as pd
import numpy as np

# Load CSV file
csv_file_path = 'Data/Flanders_dataset/P6269 - 1.50 Aanlevering DMK - E - Kwartierwaarden.csv'  # Provide the path to your CSV file
data_from_csv = pd.read_csv(csv_file_path, header=None)

# Calculate the sum of the 3rd and 4th values for each row
values_to_fill = float(data_from_csv.iloc[:, 2]) + float(data_from_csv.iloc[:, 3])

# Create date range for the year 2021
start_date = pd.Timestamp('2021-01-01')
end_date = pd.Timestamp('2021-12-31')
date_range = pd.date_range(start_date, end_date, freq='D')

# Generate multi-index with meterid and date
meter_ids = np.arange(1, 101)
index = pd.MultiIndex.from_product([meter_ids, date_range], names=['meterID', 'date'])

# Create 96 columns for each 15-minute interval of the day
columns = pd.date_range('2021-01-01', periods=96, freq='15min')

# Create DataFrame with random values
data = np.random.randint(1, 101, size=(len(index), len(columns)))
df = pd.DataFrame(data, index=index, columns=columns)

# Iterate over the DataFrame and fill it with values from the CSV
for i, (index, row) in enumerate(df.iterrows()):
    df.loc[index] = values_to_fill[i % len(values_to_fill)]

df.head()
```

```python
import pandas as pd

# Load the CSV file
csv_file_path = 'Data/Flanders_dataset/P6269 - 1.50 Aanlevering DMK - E - Kwartierwaarden.csv'  # Provide the path to your CSV file
df = pd.read_csv(csv_file_path)

# Group by 'EAN_ID' and sum the 'VolumeAfname_TH_kWh' and 'VolumeInjectie_TH_kWh' columns
grouped_df = df.groupby('EAN_ID').agg({'VolumeAfname_TH_kWh': 'sum', 'VolumeInjectie_TH_kWh': 'sum'})

# Calculate the yearly consumption
grouped_df['yearly_consumption'] = grouped_df['VolumeAfname_TH_kWh'] - grouped_df['VolumeInjectie_TH_kWh']
grouped_df['hasSolar'] = grouped_df['VolumeInjectie_TH_kWh'] == 0
grouped_df['injected'] = grouped_df['VolumeInjectie_TH_kWh']
grouped_df['consumed'] = grouped_df['VolumeAfname_TH_kWh']

# Drop unnecessary columns
grouped_df.drop(['VolumeAfname_TH_kWh', 'VolumeInjectie_TH_kWh'], axis=1, inplace=True)
grouped_df = grouped_df.rename_axis('meterID')

grouped_df.head()
```

```python
merged_attributes = repeated_df.merge(grouped_df,on=['meterID'], how="left")
```
